```python
import matplotlib.pyplot as plt
import matplotlib.colors as cores
import numpy as np
```

```python
def deamostras(x, factor):
    return [x[i] for i in range(len(x)) if i % factor == 0]
```

```python
n = np.arange(-50, 51)
x = np.sin(.125 * np.pi * n)

plt.figure(figsize=(16, 4))
axes = plt.gca()
axes.set_xlim([-55, 55])
axes.set_ylim([-1.1, 1.1])
markerline, stemlines, baseline = plt.stem(n, x, '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()

plt.figure(figsize=(16, 4))
axes = plt.gca()
axes.set_xlim([-55, 55])
axes.set_ylim([-1.1, 1.1])
markerline, stemlines, baseline = plt.stem(deamostras(n, 4), deamostras(x, 4), '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))

plt.show()
```
