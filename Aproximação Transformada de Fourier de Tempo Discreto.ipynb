```python
import matplotlib.pyplot as plt
import matplotlib.colors as cores
import numpy as np

M = 128
n = np.arange(11)
k = 4 * np.arange(-M/2, M/2 -1)
x = (.9 * np.exp(1j * np.pi/3)) ** n

def dtft_approx(x, n, w):
    return np.matmul(x, np.exp((np.pi/M) * (np.outer(n, k)) * -1j))

X = dtft_approx(x, n, k)

# print(X)

plt.figure(figsize=(16, 4))
markerline, stemlines, baseline = plt.stem(k, np.absolute(X), '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()

plt.figure(figsize=(16, 4))
markerline, stemlines, baseline = plt.stem(k, np.angle(X), '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()
```
