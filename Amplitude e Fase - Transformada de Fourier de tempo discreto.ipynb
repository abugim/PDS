```python
import matplotlib.pyplot as plt
import matplotlib.colors as cores
import numpy as np

w = np.linspace(-np.pi, np.pi, 100)
X = 1/ (1 - .7 * np.exp(1j * w))

plt.figure(figsize=(16, 4))
markerline, stemlines, baseline = plt.stem(w, np.absolute(X), '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()

plt.figure(figsize=(16, 4))
markerline, stemlines, baseline = plt.stem(w, np.angle(X), '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()

plt.figure(figsize=(16, 4))
markerline, stemlines, baseline = plt.stem(w, np.real(X), '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()

plt.figure(figsize=(16, 4))
markerline, stemlines, baseline = plt.stem(w, np.imag(X), '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()
```
