```python
import matplotlib.pyplot as plt
import matplotlib.colors as cores
import numpy as np
```

```python
x = [.9 ** n for n in np.arange(0, 21)]
y = [.8 ** -n for n in np.arange(-20, 1)]

r_xx = np.convolve(x, x[::-1])
r_xy = np.convolve(x, y[::-1])
```

```python
plt.figure(figsize=(16, 4))
markerline, stemlines, baseline = plt.stem(np.arange(0, 21), x, '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()

x_r_xy = np.arange(-20, 21)
plt.figure(figsize=(16, 4))
markerline, stemlines, baseline = plt.stem(np.arange(-20, 1), y, '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()
```

```python
x_r_xx = np.arange(-20, 21)
plt.figure(figsize=(16, 4))
markerline, stemlines, baseline = plt.stem(x_r_xx, r_xx, '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()

x_r_xy = np.arange(0, 41)
plt.figure(figsize=(16, 4))
markerline, stemlines, baseline = plt.stem(x_r_xy, r_xy, '--k', markerfmt='b.')
plt.setp(baseline, 'color', 'black', 'linewidth', 2)
plt.setp(stemlines, 'color', cores.to_rgba('black', .5))
plt.setp(markerline, 'color', cores.to_rgba(plt.getp(markerline, 'color'), .5))
plt.show()
```
